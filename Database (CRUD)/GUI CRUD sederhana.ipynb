```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

window = Tk()
window.title('CRUD Sederhana')
window.geometry('925x400+200+100')
window.configure(bg='#fff')
window.resizable(False, False)

def clear():
    txt_no.delete(0, 'end')
    txt_nama.delete(0, 'end')

def create():
    No = txt_no.get()
    Nama = txt_nama.get()
    try:
        cursor.execute(f"INSERT INTO data (No, Nama) VALUES('{No}', '{Nama}')")
        db.commit()
        tampil()
    except:
        messagebox.showerror('Informasi','No sudah dipakai, gunakan No yang lain !!!')

def update():
    No = txt_no.get()
    Nama = txt_nama.get()
    cursor.execute(f"UPDATE data SET nama='{Nama}' WHERE No='{No}'")
    db.commit()
    tampil()

def delete():
    No = txt_no.get()
    Nama = txt_nama.get()
    cursor.execute(f"DELETE FROM data WHERE No='{No}' AND Nama='{Nama}'")
    db.commit()
    tampil()
    
def tampil():
    for item in table.get_children():
        table.delete(item)
    cursor.execute("SELECT * FROM data")
    data = cursor.fetchall()
    for row in data:
        table.insert("", "end", values=row)
        

#Pembagian frame untuk mempermudah design
frame_atas = Frame(window, width=925, height=150, bg='grey')
frame_atas.place(x=0,y=0)

frame_bawah = Frame(window, width=925, height=350, bg='blue')
frame_bawah.place(x=0,y=150)


#Database
host='localhost'
user="root"
password=""

import mysql.connector as mysql

db = mysql.connect(
    host=host,user=user, password=password
)

cursor = db.cursor() 
db.database="latihan"


#Attribut button, entry, label
lbl_no = Label(frame_atas, text='No', fg='black', bg='white', font=('Microsoft Yahei UI Light', 20, 'bold'))
lbl_no.place(x=60, y=20)
txt_no = Entry(frame_atas, width=25, fg='black',border=1, bg='white',font=('Microsoft Yahei UI Light', 20))
txt_no.place(x=180, y=20)

lbl_nama = Label(frame_atas, text='Nama', fg='black', bg='white', font=('Microsoft Yahei UI Light', 20, 'bold'))
lbl_nama.place(x=60, y=90)
txt_nama = Entry(frame_atas, width=25, fg='black',border=1, bg='white',font=('Microsoft Yahei UI Light', 20))
txt_nama.place(x=180, y=90)

btn_create = Button(frame_atas, width=15, pady=7,text='Create', bg='white', fg='blue', command=create)
btn_create.place(x=590, y=20)

btn_update = Button(frame_atas, width=15, pady=7,text='Update', bg='white', fg='blue', command=update)
btn_update.place(x=590, y=90)

btn_clear = Button(frame_atas, width=15, pady=7,text='Clear', bg='white', fg='red', command=clear)
btn_clear.place(x=740, y=20)

btn_delete = Button(frame_atas, width=15, pady=7,text='Delete', bg='white', fg='red', command=delete)
btn_delete.place(x=740, y=90)

#TreeView
table = ttk.Treeview(frame_bawah, column=("No","Nama"))
table.heading('No',text='No')
table.heading('Nama', text='Nama')
table.column('No', width=150)
table.column('Nama', width=700)
table['show'] = 'headings'
table.place(x=35, y=30, height=120)

tampil()

def displaySelectedItem(a):
    
    #Clear Entry
    txt_no.delete(0, END)
    txt_nama.delete(0, END)
    
    selectedItem = table.selection()
    
    if not selectedItem:
        return
    
    txt_no.insert(0, table.item(selectedItem)['values'][0])
    txt_nama.insert(0, table.item(selectedItem)['values'][1])

table.bind('<<TreeviewSelect>>', displaySelectedItem)

window.mainloop()
```
